# 数据预处理：获取星期和转换通话时间单位等

```python
import time,datetime
def string2timestamp(x):
    strValue = str(x['开始日期'])
    #print(strValue)
    try:        
        d = datetime.datetime.strptime(strValue, "%Y-%m-%d %H:%M:%S.%f")
        t = d.timetuple()
        timeStamp = int(time.mktime(t))
        timeStamp = int(float(str(timeStamp) + str("%06d" % d.microsecond))/1000000)   
    except ValueError as e:
        d = datetime.datetime.strptime(strValue, "%Y-%m-%d %H:%M:%S")
        t = d.timetuple()
        timeStamp = int(time.mktime(t))
        timeStamp = int(float(str(timeStamp) + str("%06d" % d.microsecond))/1000000)
    date = timeStamp
    ltime = time.localtime(int(date))    
    dateymd = time.strftime("%Y-%m-%d", ltime)
    week = datetime.datetime.strptime(dateymd, "%Y-%m-%d").weekday()
    return week

data['week'] = data.apply(lambda x:string2timestamp(x),axis=1)
data
```

```python
#数据清洗 去除含'-'的行
drop_index = []
transform_list = ['平均排队等待时长','最长排队等待时长','挂断者的平均等待时长','挂断者的最长等待时长','平均通话时间','最长通话时间',
               '回电的平均排队时长','回电的平均排队等待时间','回电的最长排队时间','回电的最长等待时间',
               '平均通话时长(接入+回电)','最长的通话时间(接入+回电)']
for i in range(data.shape[0]):
    for col in transform_list:
        if isinstance(data.iloc[i][col],(str,float)):
            drop_index.append(i)
drop_index = list(set(drop_index))
print(len(drop_index))
new_data = data[~data.index.isin(drop_index)]

from datetime import datetime, date, time, timedelta
for col in transform_list:
    new_data[col] = [int((datetime.combine(date.min, i) - datetime.min).total_seconds()) for i in new_data[col]]
new_data.head(20)
```

# 2.1 每组一个月的呼叫总数 --南丁格尔玫瑰图 (python3.7版本)

```python
#每组一个月的呼叫总数
all_incall = data[['分配队列(电话内容)','总通话数']].groupby(['分配队列(电话内容)']).sum().reset_index()
all_come = sum(list(all_incall['总通话数']))
all_incall['percentage'] = 100 * all_incall['总通话数'] / all_come
all_incall['percentage'] = [round(i,2) for i in all_incall['percentage']]
all_incall.to_csv('每组一个月的呼叫总数占比.csv',index=False)
```

```python
all_incall = all_incall.sort_values(by=['percentage'],ascending=False)
all_incall
```

```python
import pandas as pd
from collections import Counter
from pyecharts.charts import Pie
from pyecharts import options as opts

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)


#南丁格尔图
group = list(all_incall['分配队列(电话内容)'])
num = list(all_incall['percentage']) 
color_series = ['#FAE927', '#E9E416', '#C9DA36', '#9ECB3C', '#6DBC49',
                '#37B44E', '#3DBA78', '#14ADCF', '#209AC9', '#1E91CA',
                '#2C6BA0', '#2B55A1', '#2D3D8E', '#44388E', '#6A368B'] #颜色设置
df = pd.DataFrame({'group': group, 'num': num})
df.sort_values(by='num', ascending=False, inplace=True)
v = df['group'].values.tolist()
d = df['num'].values.tolist()
pie1 = Pie(init_opts=opts.InitOpts(width='1350px', height='750px')) #画南丁格尔图设置

# 设置颜色
#pie1.set_colors(color_series)
# 添加数据，设置饼图的半径，是否展示成南丁格尔图
pie1.add("", [list(z) for z in zip(v, d)],
         radius=["30%", "135%"],
         center=["50%", "65%"],
         rosetype="area"
         )
# 设置全局配置项
pie1.set_global_opts(title_opts=opts.TitleOpts(title='The ratio of the monthly call volume of each group to the total call volume of the group'),
                     legend_opts=opts.LegendOpts(is_show=False),
                     toolbox_opts=opts.ToolboxOpts())
# 设置系列配置项
pie1.set_series_opts(label_opts=opts.LabelOpts(is_show=True, position="inside", font_size=12,
                                               formatter="{b}:{c}%", font_style="italic",
                                               font_weight="bold", font_family="Microsoft YaHei"
                                               ),
                     )
# 生成html文档
pie1.render('南丁格尔玫瑰图_通话量占比.html')
```
